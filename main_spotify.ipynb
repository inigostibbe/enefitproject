```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Your client ID and client secret from Spotify Developer Dashboard
client_id = ''
client_secret = ''

# Setup authentication
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

def get_artist_top_tracks(artist_name):
    # Search for the artist by name
    result = sp.search(q='artist:' + artist_name, type='artist')
    if result['artists']['items']:
        # Get the first artist from the search results
        artist = result['artists']['items'][0]
        print(f"Found artist: {artist['name']}")

        # Get top tracks for the found artist
        top_tracks = sp.artist_top_tracks(artist['id'])
        print("\nTop tracks:")
        for i, track in enumerate(top_tracks['tracks'], start=1):
            print(f"{i}. {track['name']} - {track['album']['name']}")

    else:
        print("Artist not found.")

# Example usage
get_artist_top_tracks('Danny Chaska')

```
